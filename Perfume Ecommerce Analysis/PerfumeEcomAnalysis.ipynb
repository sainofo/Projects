```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# clustering
mp_df = pd.read_csv("ebay_mens_perfume_clean.csv", parse_dates=["lastUpdated"])
mp_df['days_since_update'] = (pd.Timestamp.now() - mp_df['lastUpdated']).dt.days

features = ['price', 'available', 'sold', 'days_since_update']
X = mp_df[features].fillna(0)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

inertia = []
k_range = range(1, 10)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```
